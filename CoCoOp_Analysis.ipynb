```python
import torch
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
```

```python
prompts = torch.load("prompt_outputs/cocoop_prompts.pt")
flat = prompts.mean(dim=1).numpy()
```

```python
prompts.shape
```

```python
plt.hist(flat.flatten(), bins=100)
```

```python
!pip install umap-learn
```

```python
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

# Apply PCA to reduce dimensionality from 512 to 50
pca = PCA(n_components=50)
pca_embeddings = pca.fit_transform(flat)

# Apply t-SNE to the PCA-reduced data
tsne = TSNE(n_components=2, perplexity=5, init="random", n_iter=5000, random_state=42)
tsne_embeddings = tsne.fit_transform(pca_embeddings)

# Plot the t-SNE visualization
plt.scatter(tsne_embeddings[:, 0], tsne_embeddings[:, 1], s=10)
plt.title("t-SNE of PCA-Reduced CoCoOp Prompts")
plt.show()
```

```python
import umap
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

# Apply PCA to reduce dimensionality from 512 to 50
pca = PCA(n_components=50)
pca_embeddings = pca.fit_transform(flat)
```

```python
# Apply UMAP to the PCA-reduced embeddings
umap_model = umap.UMAP(n_components=2, n_neighbors=10, min_dist=0.1, random_state=42)
umap_embeddings = umap_model.fit_transform(pca_embeddings)
```

```python
colors = pca_embeddings[:, 0]  # First PCA component as color

plt.scatter(umap_embeddings[:, 0], umap_embeddings[:, 1], c=colors, cmap="viridis", s=10)
plt.title("UMAP of PCA-Reduced CoCoOp Prompts")
plt.colorbar()
plt.show()
```

```python
from sklearn.cluster import KMeans

num_clusters = 102  # Number of clusters (adjust as needed)
kmeans = KMeans(n_clusters=num_clusters, random_state=42, n_init=10)
cluster_labels = kmeans.fit_predict(pca_embeddings)

plt.scatter(umap_embeddings[:, 0], umap_embeddings[:, 1], c=cluster_labels, cmap="jet", s=10)
plt.title(f"UMAP of CoCoOp Prompts with {num_clusters} Clusters")
plt.colorbar()
plt.show()
```
